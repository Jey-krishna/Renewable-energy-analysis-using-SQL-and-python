```python
df['Initial_Investment_USD'] = df['Initial_Investment_USD'].apply(lambda x: '{:.0f}'.format(x) if pd.notnull(x) else x)
df['Financial_Incentives_USD'] = df['Financial_Incentives_USD'].apply(lambda x: '{:.0f}'.format(x) if pd.notnull(x) else x)

```

```python
df
```

```python
df['Installed_Capacity_MW'] = df['Installed_Capacity_MW'].apply(lambda x: '{:.2f}'.format(x) if pd.notnull(x) else x)
df['Energy_Production_MWh'] = df['Energy_Production_MWh'].apply(lambda x: '{:.2f}'.format(x) if pd.notnull(x) else x)
df['Energy_Consumption_MWh'] = df['Energy_Consumption_MWh'].apply(lambda x: '{:.2f}'.format(x) if pd.notnull(x) else x)
df['Energy_Storage_Capacity_MWh'] = df['Energy_Storage_Capacity_MWh'].apply(lambda x: '{:.2f}'.format(x) if pd.notnull(x) else x)
df['Storage_Efficiency_Percentage'] = df['Storage_Efficiency_Percentage'].apply(lambda x: '{:.2f}'.format(x) if pd.notnull(x) else x)
df['GHG_Emission_Reduction_tCO2e'] = df['GHG_Emission_Reduction_tCO2e'].apply(lambda x: '{:.2f}'.format(x) if pd.notnull(x) else x)
df['Air_Pollution_Reduction_Index'] = df['Air_Pollution_Reduction_Index'].apply(lambda x: '{:.2f}'.format(x) if pd.notnull(x) else x)



```

```python
df
```

```python
df['Type_of_Renewable_Energy'].replace(1, 'Solar')
```

```python
df['Type_of_Renewable_Energy'] = df['Type_of_Renewable_Energy'].replace(
    {
        1 : 'Solar',
        2 : 'Wind',
        3 : 'Hydroelectric',
        4 : 'Geothermal',
        5 : "Biomass",
        6 : 'Tidal',
        7 : 'Wave'
    }
)
```

```python
df['Grid_Integration_Level'] = df['Grid_Integration_Level'].replace(
    {
        1 : 'Fully Integrated',
        2 : 'Partially Integrated',
        3 : 'Minimal Integration',
        4 : 'Isolated Microgrid',
    }
)
```

```python
df['Funding_Sources'].unique()
```

```python
df['Funding_Sources'] = df['Funding_Sources'].replace(
    {
        1 : 'Government',
        2 : 'Private',
        3 : 'Public-Private Partnership'
    }
)
```

```python
df.rename(columns= {'Funding_Sources\t' : 'Funding_Sources'}, inplace=True)
```
